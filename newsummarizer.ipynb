```python
from transformers import pipeline
import gradio as gr

# Load the summarization pipeline
summarizer = pipeline("summarization", model="t5-small", tokenizer="t5-small")

# Define summarization function
def summarize_article(article):
    if not article.strip():
        return "Please enter a news article."

    # T5 expects input like: "summarize: <text>"
    input_text = "summarize: " + article.strip().replace("\n", " ")
    summary = summarizer(input_text, max_length=100, min_length=30, do_sample=False)
    return summary[0]['summary_text']

# Create Gradio UI
demo = gr.Interface(
    fn=summarize_article,
    inputs=gr.Textbox(lines=15, placeholder="Paste your news article here...", label="News Article"),
    outputs=gr.Textbox(label="Summary"),
    title="📰 News Summarizer",
    description="Paste a news article to generate a concise summary using T5 Transformer."
)

demo.launch(share=True, debug=True)
```
